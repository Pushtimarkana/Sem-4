```python
import matplotlib.pyplot as plt

x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]

# write a code to generate below graph

plt.plot(x,cxMarks,ls="--",lw=1,c="r",marker=">",label="CX")
plt.plot(x,cyMarks,ls=":",lw=1,c="b",marker="o",label="CY")
plt.xlabel("Roll No")
plt.ylabel("CPI")
plt.annotate("Lowest CPI", xy=[8,1.5])
plt.legend()
plt.show()
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
values= [82,50,60,36,92]
l= ["Maths","Python","Flutter","DBMS","SE"]
x = ["c","b","m","g","r"]
bars=plt.bar(l,values,color=x,width=0.5)
# for i in bars:
#     yc = i.get_height()
#     plt.text(i.get_x(), yc+2, f"{yc}")

for i in bars:
    plt.text(
        i.get_x()+i.get_width()/2,
        i.get_height(),
        i.get_height(),
        ha="center",
        va="bottom"
    )
plt.show()
```
